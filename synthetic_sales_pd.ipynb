```python
df['Value_new']=df['Value'].astype(float)
df.head()
```

data aggregating and grouping 


```python
grouped__mean=df.groupby('Product')['Value'].mean()
print(grouped__mean)
```

```python
grouped_sum=df.groupby(['Product','Region'])['Value'].sum()
print(grouped_sum)
```

aggrgate multiple function

```python
grouped_agg=df.groupby(['Product','Region'])['Value'].agg(['mean','sum','count'])  
grouped_agg
```

merging and joining dataframe


```python

##like a vein 
df1=pd.DataFrame({'Key':['A','B','C'],'Value1':[1,2,3]})
df2=pd.DataFrame({'Key':['A','B','D'],'Value2':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
#Merge Dataframe on 'Key column'
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on="Key",how="inner")
```

```python
pd.merge(df1,df2,on="Key",how="outer")
```

```python
pd.merge(df1,df2,on="Key",how="left")
```

```python
pd.merge(df1,df2,on="Key",how="right")
```
